# Bar Plots
Bar plots are most commonly used to plot the frequencey of the possible values of a <span style="color:red;">categorical</span> variable.

They are very useful for displaying the class/feature distributions are in our data.

<big>matplotlib is a common library of choice to generate plots and figures.</big><br>
provides high-performance, easy-to-use data structures and data analysis tools.<br>
One of the most commonly used tools in the AI and Data Science field.

Matplotlib is a versatile Python library for creating a wide range of visualizations, favored for its simplicity for generating those visuals. 

```python
import matplotlib.pyplot as plt
```

<big>For our use case, we want to plot the frequencey of each class/category as a bar for one of our features.</big>

`pandas.DataFrame.value_counts()` will be very useful for us to quickly retrieve the categories and their respective frequency.

```python
def make_bar_plot(dataframe, column_name):
    """shows a barplot of a categorical feature (column).

        Parameters
        ----------
        dataframe : pandas.DataFrame, required
            Dataframe that contains all the data

        column_name : str, required
            name of the column that contains the values of the categorical feature.
    """
    # Counting the occurrence of each category in column "A"
    category_counts = dataframe[column_name].value_counts()
    
    # turn the index of the categories into strings instead of integers
    category_counts.index = category_counts.index.astype(str)
    
    # Extracting category names and counts
    categories = category_counts.index
    counts = category_counts.values

    # Plotting a bar plot
    plt.bar(categories, counts)

    # Adding labels and title
    plt.xlabel('Categories')
    plt.ylabel('Count')
    plt.title(f'Occurrences of Categories in Column \'{column_name}\'')

    # Rotating x-axis labels for better readability
    plt.xticks(rotation=45)

    # Showing the plot
    plt.show()
```

```python
make_bar_plot(df, "Pclass")
```

# Histograms
Histograms are useful for plotting the probability distribution of a numerical variable 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def make_hist_plot(df, column_name):
    """Plots the distribution of a specified column in the DataFrame using a histogram.

        Parameters
        ----------
        df : pandas.DataFrame
            Dataframe that contains all the data

        column_name : str
            name of the column that contains the values of the feature.
    """
    # Check if column exists in DataFrame
    if column_name not in df.columns:
        print(f"Column '{column_name}' not found in DataFrame.")
        return
    
    # Plotting the histogram
    plt.hist(df[column_name], bins='auto', color='skyblue', alpha=0.7, rwidth=0.85, density=True)
    
    # Plot aesthetics
    plt.title(f'Distribution of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.grid(axis='y', alpha=0.75)

    plt.show()
```

```python
# Plotting the distribution of the 'age' column
make_hist_plot(df, 'Age')
```

## Use Case #1
For comparing and visualizing the relationship between two variables

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def make_scatter_plot(df, x_column, y_column=None, fit_line=True):
    """
    Plots a scatter plot of the values in the specified columns of the DataFrame.

    Parameters:
    - df: pandas.DataFrame containing the data.
    - x_column: str, the name of the column to use for the x-axis.
    - y_column: str, optional, the name of the column to use for the y-axis. If not provided, the index of the DataFrame will be used.
    - fit_line: bool, optional, whether to draw a best fit line or not. Default is True.
    """
    # Check if columns exist in DataFrame
    if x_column not in df.columns:
        print(f"Column '{x_column}' not found in DataFrame.")
        return
    if y_column is not None and y_column not in df.columns:
        print(f"Column '{y_column}' not found in DataFrame.")
        return

    # Extract values from the specified columns
    x_values = df[x_column]
    y_values = df[y_column] if y_column else df.index

    # Plot scatter plot
    plt.scatter(x_values, y_values, label='Data points', color='skyblue')

    # Plot best fit line if fit_line is True
    if fit_line:
        # Calculate slope and intercept of the best fit line
        slope, intercept = np.polyfit(x_values, y_values, deg=1)

        # Generate y values for the best fit line
        fit_line_values = slope * x_values + intercept

        # Plot the best fit line
        plt.plot(x_values, fit_line_values, label='Fit line', color='red')

    # Plot aesthetics
    plt.title(f'Scatter Plot of {x_column} vs {y_column if y_column else "Index"}')
    plt.xlabel(x_column)
    plt.ylabel(y_column if y_column else 'Index')
    plt.legend()

    plt.show()
```

```python
sample_df = pd.DataFrame({'data': np.random.normal(0, 1, 100)})

# Plotting scatter plot with fit line for the 'data_column'
make_scatter_plot(sample_df, 'data')
```

## Use Case #2
For Visualizing groups and clusters of data.

```python
import pandas as pd
from matplotlib import pyplot as plt

PATH = '../Datasets/Iris/iris.data'
iris_data = pd.read_csv(PATH,
                        names=["sepal_length","sepal_width","petal_length","petal_width","class"]
                       )
iris_data
```

```python
def plot_clusters(df, categorical_column, X, Y):
    """
    Creates a scatter plot using the values of two columns (X, Y) and colors the points based on categorical
    column value.

    Parameters:
    - df: pandas.DataFrame containing the data.
    - categorical_column: str, name of the column for the categorical feature.
    - X: str, the name of the column to use for the x-axis.
    - Y: str,the name of the column to use for the y-axis.
    """
    # manipulate the size (resolution) of the output figure/image.
    # this line MUST be put before the declaration of the figure type.
    plt.figure(figsize=(8, 6))
    
    # loop through the classes of the categorical feature and 
    #    plot the points of each class with a unique color.
    # Values are retrieved using a pandas conditional selection method (similar behaviour to .loc method).
    for class_name in df[categorical_column].unique():
        plt.scatter(df[df[categorical_column] == class_name][X],
                    df[df[categorical_column] == class_name][Y],
                    label=class_name, s=50)
    
    plt.xlabel(X)
    plt.ylabel(Y)
    plt.title('Scatter Plot of Sepal Length vs Sepal Width (Clustered by Species)')
    plt.legend()
    plt.grid(True)
    
    plt.show()
```

```python
plot_clusters(df=iris_data,
              categorical_column = 'class',
              X = 'sepal_length',
              Y = 'sepal_width',)
```
