```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# 1. Load data
train_data = pd.read_csv('/kaggle/input/just-ai-249-machine-learning-lab-competition/spambase_train.csv')
test_data = pd.read_csv('/kaggle/input/just-ai-249-machine-learning-lab-competition/spambase_test.csv')

# Check column names
print("Train data columns:", train_data.columns)
print("Test data columns:", test_data.columns)

# 2. Prepare data
if 'spam' in train_data.columns:
    X_train = train_data.drop(columns=['spam', 'ID'])  # Remove 'ID' column from training data
    y_train = train_data['spam']
else:
    raise KeyError("Column 'spam' not found in train data")

if 'ID' in test_data.columns:
    X_test = test_data.drop(columns=['ID'])
    test_ids = test_data['ID']
else:
    raise KeyError("Column 'ID' not found in test data")

# 3. Build model with speed improvements
clf = RandomForestClassifier(n_estimators=50, max_depth=10, min_samples_split=10, n_jobs=-1, random_state=42)
clf.fit(X_train, y_train)

# 4. Predict on test data
predictions = clf.predict(X_test)

# 5. Save results
submission = pd.DataFrame({
    'ID': test_ids,
    'spam': predictions
})

submission.to_csv('motasem.csv', index=False)
```
