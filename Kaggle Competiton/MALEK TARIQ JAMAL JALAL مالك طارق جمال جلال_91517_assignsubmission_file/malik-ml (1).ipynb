Here we downloaded the evidence and prepared and we also deleted the ID and Spam pillars from the training data
 And then we kept the spam pillar as lable
And then I did it if there are empty or not present values ​​to improve its quality

```python
import pandas as pd

train_data = pd.read_csv('/kaggle/input/just-ai-249-machine-learning-lab-competition/spambase_train.csv')
test_data = pd.read_csv('/kaggle/input/just-ai-249-machine-learning-lab-competition/spambase_test.csv')

X_train = train_data.drop(columns=['ID', 'spam'])
y_train = train_data['spam']

X_train.fillna(X_train.median(), inplace=True)
```

And here I summoned the function to divide the data into Traineing data and testing data

```python
from sklearn.model_selection import train_test_split

X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

And here I prepared and prepared the logistical slope and its training
And I also made Pipeline to scale the data
 And I got here too ,`param_grid` specifies the hyperparameters for `LogisticRegression` to tune: `C` controls the regularization strength, and `penalty` determines the type of regularization (l1 or l2). GridSearchCV uses these to find the best model 
GridSearchcv: You search for the best collection of parameters and evaluate the model with the F1
grid_search.fit : And this function is training the model
Then it stores the best result in the best_model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(solver='liblinear'))
])

param_grid = {
    'model__C': [0.01, 0.1, 1, 10, 100],
    'model__penalty': ['l1', 'l2']
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=2)
grid_search.fit(X_train_split, y_train_split)

best_model = grid_search.best_estimator_
```

I wrote this symbol to evaluate the performance of the model, and here the trained model also performs predictions on (X_VAL)
 It turns predictions into categories (0 or 1) based on a 0.5 threshold. If the expected value is greater than or equal to 0.5, then it is considered 1 (spam mail), otherwise it is considered 0 (not spam mail)
 Duration: You can calculate the correct prediction rate
 F1: Calculate F1 degree

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score

y_val_pred = best_model.predict(X_val)
y_val_pred_class = (y_val_pred >= 0.5).astype(int)

accuracy = accuracy_score(y_val, y_val_pred_class)
f1 = f1_score(y_val, y_val_pred_class)
report = classification_report(y_val, y_val_pred_class)

print(f'Accuracy: {accuracy:.2f}')
print(f'F1 Score: {f1:.2f}')
print('Classification Report:')
print(report)
```

Here's the first thing I did. I have delete the ID pillar of data to choose because I do not need it to predict
Then I used the Best Modle to make predictions on (X Test)
We Convert Predists InTo Categories (0 or 1) Using a 0.5 Thresthold. If the predictive Value is Greater Than or Equal to 0.5, then it is ConsiderD 1 (spam), and if it is less for 0.5, it is considored 0 (not spam).
The Csv File Creates Predations and Definition Numbers (ID) to save the Results.

```python
X_test = test_data.drop(columns=['ID'])
X_test.fillna(X_test.median(), inplace=True)

y_test_pred = best_model.predict(X_test)
y_test_pred_class = (y_test_pred >= 0.5).astype(int)

submission_df = pd.DataFrame({'ID': test_data['ID'], 'spam': y_test_pred_class})
submission_df.to_csv('malik11.csv', index=False)
```
