```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report

training_df = pd.read_csv('/kaggle/input/just-ai-249-machine-learning-lab-competition/spambase_train.csv')

testing_df = pd.read_csv('/kaggle/input/just-ai-249-machine-learning-lab-competition/spambase_test.csv')


features_training = training_df.drop(columns=['spam', 'ID'])
labels_training = training_df['spam']
```

# splits the data into two parts:1- training part and 2- validation part 
**to ensure that the model is trained on one part of the data, and evaluated on another part that it has not seen before. traning data in 80% from data and testing in 20% from data and I used the train_test_split function for this .**
# In the variable rf_classifier
**I used a function of the RandomForestClassifier algorithm and assigned the decision tree to be 150 in the model. I used it before 100 and it was somewhat good, but I preferred this number, and when I used a larger number, there was somewhat improvement, but I needed more time, memory, and my laptop. Don't put up with it
 random_state=42: I used this number using some references, and they described it as good. I saw this in my model, as it specializes in making a random choice every time, but with a fixed number and number, in order to ensure that the change in performance is not due to rerunning the code, but due to other conditions that I have changed.**

 

```python
features_training_split, features_validation_split, labels_training_split, labels_validation_split = train_test_split(features_training, labels_training, test_size=0.2, random_state=42)


 rf_classifier = RandomForestClassifier(n_estimators=150, random_state=58)
```

**1- I used the fit function to learn the model: the model learns the relationship between features and labels from the training data**

**2- Using the predict function, I used it to predict the labels for the validation set. So I can see how it works on new data that I haven't seen before**

**3- Calculate accuracy to evaluate model performance using the accuracy_score function**

**4- Calculate the precision/recall ratio (F1 Score) to evaluate the model performance. Using the f1_score function 
(f1 score = 2* (precision*recall/ precision+recall) )**

```python

rf_classifier.fit(features_training_split, labels_training_split) # 1

predicted_labels_validation = rf_classifier.predict(features_validation_split) # 2


accuracy = accuracy_score(labels_validation_split, predicted_labels_validation) # 3
f1 = f1_score(labels_validation_split, predicted_labels_validation)# 4
# print 
print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
print('Classification Report:')
print(classification_report(labels_validation_split, predicted_labels_validation))
```

**1-** **For the X_test_data variable, it prepares the features of the test data by removing the ID column, because it is not part of the features that the model needs to predict, so I removed it.**

**2-** **For the variable predicted_labels_test Now I will predict the labels (spam or not spam) for the new samples in the test data set.**

```python
test_ids = testing_df['ID'] 
X_test_data = testing_df.drop(columns=['ID']) # 1

predicted_labels_test = rf_classifier.predict(X_test_data) # 2 

submission_data = pd.DataFrame({'ID': test_ids, 'spam': predicted_labels_test})

print("Submission DataFrame:")
print(submission_data.head())

submission_data.to_csv('adnan_ML_Final_3.csv', index=False)
```
