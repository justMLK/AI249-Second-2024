# K-Fold Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

from numpy import mean
from numpy import std
```

```python
config = {
    "PATH": '../Datasets/Iris/iris.data',
    
    "features": ["sepal_length","sepal_width","petal_length","petal_width"],
    "target": "class",
    
    "seed": 100, # random_seed
    "test_size": 0.3, # try 0.2, 0.3, 0.5
    
    
    #________Hyperparameters________#
    "kernel": "linear", # kernel options: {‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’}
    "C": 1.0,
    "gamma": 1.0, #gamma{‘scale’, ‘auto’}
    
    "max_iter": -1, # Hard limit on iterations, -1 means no limit until convergence/no change   
}

column_names = ["sepal_length","sepal_width","petal_length","petal_width","class"]
iris_data = pd.read_csv(config["PATH"],
                        names=column_names
                       )
```

```python
def kfold_cross_validation(df, K=5, scoring="accuracy"):
    """
    options for scoring: 
        accuracy, f1, f1_micro, f1_macro, precision, recall
    
    """
    feature_columns = config["features"]
    X = df[feature_columns].to_numpy()
    y = df[config["target"]]

    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                        test_size=config["test_size"],
                                                        random_state=config["seed"])

    model = SVC(
        kernel=config["kernel"],
        C=config["C"],
        gamma=config["gamma"],
        max_iter=config["max_iter"],
    )
    cv = KFold(n_splits=K, random_state=1, shuffle=True)
    scores = cross_val_score(model, X, y, scoring=scoring, cv=cv, n_jobs=-1)
    print(f"Performance for {K}-folds:")
    print(f"Avg. {scoring}: {mean(scores)}\nStandard Deviation: {std(scores)}")
```

```python
kfold_cross_validation(iris_data, K=5, scoring="accuracy")
```

# Grid Search for Hyper Parameter Optimization

```python
from sklearn.model_selection import GridSearchCV
```

```python
config = {
    "PATH": '../Datasets/Iris/iris.data',
    
    "features": ["sepal_length","sepal_width","petal_length","petal_width"],
    "target": "class",
    
    "seed": 100, # random_seed
    "test_size": 0.3, # try 0.2, 0.3, 0.5
    
    
    #________Hyperparameters________#
    "kernel": "linear", # kernel options: {‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’}
    "C": 1.0,
    "gamma": 1.0, #gamma{‘scale’, ‘auto’}
    
    "max_iter": -1, # Hard limit on iterations, -1 means no limit until convergence/no change   
}

column_names = ["sepal_length","sepal_width","petal_length","petal_width","class"]
iris_data = pd.read_csv(config["PATH"],
                        names=column_names
                       )
```

```python
def grid_search(df, parameters, scoring='accuracy'):
    """
    options for scoring: 
        accuracy, f1, f1_micro, f1_macro, precision, recall
    
    """
    feature_columns = config["features"]
    X = df[feature_columns].to_numpy()
    y = df[config["target"]]

    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                        test_size=config["test_size"],
                                                        random_state=config["seed"])

    model = SVC()
    grid = GridSearchCV(model, parameters, scoring = scoring)
    grid.fit(X_train, y_train)
    print(grid.best_params_)
    
    grid_predictions = grid.predict(X_test)
    report = classification_report(y_test, grid_predictions)
    print(report)
    return grid.cv_results_
```

```python
parameters = {
    'kernel':['linear', 'rbf'],
    'C':[1, 10],
    "gamma": [1.0, 3],}

search_results = grid_search(iris_data, parameters, scoring='accuracy')
```
