## Preprocessing the Dataset
Preprocessing the dataset is one of the core steps taken during the development of almost all Mahcine Learning projects.<br>
Most dataset would not be ready for analysis/training machine learning algorithms. <br>
Therefore, it is necessary that we run our dataset through a preprocessing pipeline that modifies the values in the dataset according to a rule. <br>

### Example #1 (Encoding)
<big>There are many reasons for preprocessing the values of a dataset, One of the most common ones is <span style="color:red;"><b><u>Encoding Categorical Values<u><b></span></big>

```python
def map_gender(x):
    if x == "male":
        return 0
    elif x == "female":
        return 1

# we use the pandas.series.apply() method to apply a valid preprocessing function to a specified dataset column
my_dataset['Sex'] = my_dataset['Sex'].apply(map_gender)
```

```python
my_dataset
```

### Example #2 (Binning)
<big>Another common preprocessing step is called <span style="color:red;"><b><u>Discretization<u><b></span> (Also known as Binning)</big> <br>
It is the process of turning a numerical continuous feature into a categorical feature.

```python
def map_ages(x):
    if (x < 10):
        return 0
    elif (x >= 10) & (x<20):
        return 1
    else:
        return 2        

my_dataset['Age'] = my_dataset['Age'].apply(map_ages)
```

```python
my_dataset
```

### Barplot for Categorical Features
Say we want to provide a visual representation of a categorical feature.

First of all, lets install the library `matplotlib`

```python
# !pip install matplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# For demonstration, let's create a sample DataFrame
data = {'A': ['category1', 'category2', 'category1', 'category3', 'category2', 'category1'],
        'B': [10, 20, 30, 44, 58, 61]
       }
df = pd.DataFrame(data)

# Counting the occurrence of each category in column "A"
category_counts = df['A'].value_counts()

# Extracting category names and counts
categories = category_counts.index
counts = category_counts.values

# Plotting a bar plot
plt.bar(categories, counts)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Occurrences of Categories in Column A')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Showing the plot
plt.show()
```

```python
def make_bar_plot(dataframe, column_name):
    """shows a barplot of a categorical feature (column).

        Parameters
        ----------
        dataframe : pandas.DataFrame, required
            Dataframe that contains all the data

        column_name : str, required
            name of the column that contains the values of the categorical feature.
    """
    # Counting the occurrence of each category in column "A"
    category_counts = dataframe[column_name].value_counts()
    
    # turn the index of the categories into strings instead of integers
    category_counts.index = category_counts.index.astype(str)
    
    # Extracting category names and counts
    categories = category_counts.index
    counts = category_counts.values

    # Plotting a bar plot
    plt.bar(categories, counts)

    # Adding labels and title
    plt.xlabel('Categories')
    plt.ylabel('Count')
    plt.title(f'Occurrences of Categories in Column \'{column_name}\'')

    # Rotating x-axis labels for better readability
    plt.xticks(rotation=45)

    # Showing the plot
    plt.show()
```
