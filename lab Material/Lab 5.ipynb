# Creating a testing Environment

## Dependencies

```python
import pandas as pd
import numpy as np

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore') 
```

## Utility

```python
def train_linear_sgd(df):
    
    feature_columns = config["features"]
    X = df[feature_columns].to_numpy()
    y = df[config["target"]]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=config["test_size"], random_state=config["seed"])

    model = SGDRegressor(
        loss=config["loss"], learning_rate=config["learning_rate"], eta0=config["alpha"],
        # n_iter_no_change=config["epochs"],
        verbose=0
    )


    # fit the model on the training data
    train_losses = []
    test_losses = []
    for epoch in range(config["epochs"]):
        model.partial_fit(X_train, y_train)
        
        y_hat_train = model.predict(X_train)
        y_hat_test = model.predict(X_test)
        
        train_loss = mean_squared_error(y_train, y_hat_train)
        test_loss = mean_squared_error(y_test, y_hat_test)
        
        train_losses.append(train_loss)
        test_losses.append(test_loss)

    # Plot both training and validation losses
    plt.plot(train_losses, label='Training Loss')
    plt.plot(test_losses, label='Testing Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and testing Loss over Iterations')
    plt.legend()
    plt.show()
    
    print(f"Performance on Testing Data = {test_loss}")
```

## Training

```python
config = {
    "PATH": "../Datasets/Boston Housing/housing.csv",
    "features": ['CHAS', 'NOX', 'RM'],
    "target": "MEDV",
    
    "seed": 100, # random_seed
    "test_size": 0.1,
    
    "loss": "squared_error",
    "epochs": 10000,
    
    "alpha": 0.001,
    "learning_rate": "constant",
    
}

column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM',
                'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B',
                'LSTAT', 'MEDV']
df = pd.read_csv(config["PATH"],
                 sep=r"\s+", # regular expression
                 names=column_names)
```

```python
train_linear_sgd(df)
```
