## Simplest and most important operation (matrix multiplication)

````python
# Lets define two matrices
x = [[1,2,3],
     [4,5,6],
     [7,8,9]]

y = [[1,2,3],
     [4,5,6],
     [7,8,9]]

# Initializing an empty list
z = [3*[0],
     3*[0],
     3*[0]]

# perform matrix addition
for i in range(len(z)):
    for j in range(len(z[i])):
        z[i][j] = x[i][j] + y[i][j]

for row in z:
    print(row)
````

````python
x = [[1,2,3],
     [4,5,6],
     [7,8,9]]

y = [[1,2,3],
     [4,5,6],
     [7,8,9]]

z = [3*[0],
     3*[0],
     3*[0]]

# Perform matrix multiplication
for i in range(len(x)):
    for j in range(len(y[0])):
        for k in range(len(y)):
            z[i][j] += x[i][k] * y[k][j]

for row in z:
    print(row)
````

## Norms
calculating the norm of vectors is one of the most important calculations in
data science and machine learning.

In numpy, we accomplish that using the `numpy.linalg` module

````python
"""
By default, the function uses L2-norm (also known as the 'Frobenius norm').
to specify the norm, we use the 'ord' parameter.
"""
x = np.array([1,2,3,4])
norm = np.linalg.norm(x, ord=2)
print(norm)
````

````python
"""
We can also caluclate the norm for multiple vectors
"""
x = np.random.randint(low=1, high=5, size=(3,4))
print(f"matrix:\n{x}\n")
norm = np.linalg.norm(x, ord=2, axis=0)
print(f"norms: {norm}")
````

## Preprocessing the Dataset
Preprocessing the dataset is one of the core steps taken during the development of almost all Mahcine Learning projects.<br>
Most dataset would not be ready for analysis/training machine learning algorithms. <br>
Therefore, it is necessary that we run our dataset through a preprocessing pipeline that modifies the values in the dataset according to a rule. <br>

### Example #1
<big>There are many reasons for preprocessing the values of a dataset, One of the most common ones is <span style="color:red;"><b><u>Encoding Categorical Values<u><b></span></big>

````python
def map_gender(x):
    if x == "male":
        return 0
    elif x == "female":
        return 1

# we use the pandas.series.apply() method to apply a valid preprocessing function to a specified dataset column
my_dataset['Sex'] = my_dataset['Sex'].apply(map_gender)
````

````python
my_dataset
````

### Example #2
<big>Another common preprocessing step is called <span style="color:red;"><b><u>Discretization<u><b></span> (Also known as Binning)</big> <br>
It is the process of turning a numerical continuous feature into a categorical feature.

````python
def map_ages(x):
    if (x < 10):
        return 0
    elif (x >= 10) & (x<20):
        return 1
    else:
        return 2        

my_dataset['Age'] = my_dataset['Age'].apply(map_ages)
````

````python
my_dataset
````
